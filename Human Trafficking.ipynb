```python
import pandas as pd
import numpy as np
```

```python
file18 = pd.read_csv('The Global Dataset 14 Apr 2020.csv')
file18.head()
```

```python
file19 = pd.read_csv('Total_victims_assisted.csv')
file19 = file19.rename(columns={'Year of registartion':'yearOfRegistration',
                            })
file19.head(50)
```

```python
file18.dtypes
```

```python
fileclean= file18[['yearOfRegistration','gender', 'ageBroad', 'majorityStatus', 'citizenship',  'meansOfControlConcatenated', 'typeOfExploitConcatenated', 'typeOfLabourConcatenated', 'typeOfSexConcatenated', 'isAbduction', 'RecruiterRelationship', 'CountryOfExploitation']]
fileclean.head()
```

```python
#add total of victims
merged_1 = pd.merge(left=fileclean, right=file19, how='outer')
merged_1
```

```python
merged_1.dtypes
```

```python
#working with null value
merged_1.replace('-99', np.nan, inplace=True)
merged_1.replace(-99, np.nan, inplace=True)
```

```python
#cleaning nan value
merged_1["citizenship"]= merged_1["citizenship"].fillna("Unknown")
merged_1["CountryOfExploitation"]= merged_1["CountryOfExploitation"].fillna("Unknown")
merged_1["meansOfControlConcatenated"]= merged_1["meansOfControlConcatenated"].fillna("Not specified")
merged_1['typeOfSexConcatenated'] = merged_1["typeOfSexConcatenated"].fillna("Prostitution")
merged_1['isAbduction'] = merged_1['isAbduction'].fillna(0.0)
merged_1['RecruiterRelationship'] = merged_1['RecruiterRelationship'].fillna("Not Specified")
merged_1['typeOfLabourConcatenated'] = merged_1['typeOfLabourConcatenated'].fillna("Domestic work")
merged_1['typeOfExploitConcatenated'] = merged_1['typeOfExploitConcatenated'].fillna("Sexual exploitation")

merged_1['majorityStatus'] = merged_1['majorityStatus'].fillna("Adult")
merged_1['ageBroad'] = merged_1['ageBroad'].fillna("Unknown")



```

```python
#convert countries
import country_converter as coco
converted_country = coco.convert(names=merged_1['citizenship'], to="name")
converted_country2 = coco.convert(names=merged_1['CountryOfExploitation'], to="name")
```

```python
merged_1['country_origin'] = converted_country
merged_1['country_explotation'] = converted_country2

merged_1.head()
```

```python
def notfound(x):  
    if 'not found' in x:
        return 'unknown'
    else:
        return x
    
merged_1["country_origin"] = list(map(notfound, merged_1["country_origin"]))
merged_1["country_explotation"] = list(map(notfound, merged_1["country_explotation"]))

```

```python
#reduce category
dfg4 = pd.DataFrame(merged_1['typeOfExploitConcatenated'].value_counts()).reset_index()

dfg4.columns = ['typeOfExploitConcatenated', 'count']
other_control4 = list(dfg4[dfg4['count']<160]['typeOfExploitConcatenated'])
def clean_typeex(x):
    if x in other_control4:
        return 'Sexual exploitation'
    else:
        return x
    
merged_1["typeOfExploitConcatenated"] = list(map(clean_typeex, merged_1["typeOfExploitConcatenated"]))
```

```python
merged_1.head()
```

```python
dfg3 = pd.DataFrame(merged_1['typeOfLabourConcatenated'].value_counts()).reset_index()

dfg3.columns = ['typeOfLabourConcatenated', 'count']
other_control1 = list(dfg3[dfg3['count']<70]['typeOfLabourConcatenated'])
def clean_type(x):
    if x in other_control1:
        return 'Other'
    else:
        return x
    
merged_1["typeOfLabourConcatenated"] = list(map(clean_type, merged_1["typeOfLabourConcatenated"]))
```

```python
dfg = pd.DataFrame(merged_1['meansOfControlConcatenated'].value_counts()).reset_index()

dfg.columns = ['meansOfControlConcatenated', 'count']
other_control = list(dfg[dfg['count']<180]['meansOfControlConcatenated'])
def clean_control(x):
    if x in other_control:
        return 'Not specified'
    else:
        return x
    
merged_1["meansOfControlConcatenated"] = list(map(clean_control, merged_1["meansOfControlConcatenated"]))
```

```python
dfg2 = pd.DataFrame(merged_1['RecruiterRelationship'].value_counts()).reset_index()

dfg2.columns = ['RecruiterRelationship', 'count']
other_control2 = list(dfg2[dfg2['count']<60]['RecruiterRelationship'])
def clean_recruter(x):
    if x in other_control2:
        return 'Not Specified'
    else:
        return x
    
merged_1["RecruiterRelationship"] = list(map(clean_recruter, merged_1["RecruiterRelationship"]))
```

```python
#Machine learning
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import cross_val_score
import math
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

import matplotlib.pyplot as plt 
%matplotlib inline

pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
```

```python
def OHEncoder(X):
    numericalX = X.select_dtypes(np.number)
    categoricalX = X.select_dtypes(object)

    #I'll use OneHotEncode in the categoricals
    encoder = OneHotEncoder(drop='first').fit(categoricalX)
    encoded_categorical = encoder.transform(categoricalX).toarray()
    encoded_categorical = pd.DataFrame(encoded_categorical)
    X = pd.concat([numericalX, encoded_categorical], axis = 1)
    
    return X
```

```python
def LabEncoder(X):
    numericalX = X.select_dtypes(np.number)
    categoricalX = X.select_dtypes(object)

    #encoder = LabelEncoder()#.fit_transform(categoricalX)
    encoded_categorical = categoricalX.apply(LabelEncoder().fit_transform)
    #encoded_categorical = pd.DataFrame(encoded_categorical)
    X = pd.concat([numericalX, encoded_categorical], axis = 1)
    
    return X
```

```python
def LabEncoder2(X):
    numericalX = X1.select_dtypes(np.number)
    categoricalX = X1.select_dtypes(object)

    #encoder = LabelEncoder()#.fit_transform(categoricalX)
    encoded_categorical = categoricalX.apply(LabelEncoder().fit_transform)
    #encoded_categorical = pd.DataFrame(encoded_categorical)
    X1 = pd.concat([numericalX, encoded_categorical], axis = 1)
  
    return X
```

```python
def RegModelScores(X_train_scaled, X_test_scaled, y_train, y_test):

    from sklearn.tree import DecisionTreeRegressor 
    reg = DecisionTreeRegressor(random_state = 0) 

    reg.fit(X_train_scaled,y_train)

    train_pred = reg.predict(X_train_scaled)
    test_pred = reg.predict(X_test_scaled)
    r2t = r2_score(y_train, train_pred)
    print("R2 for train:", r2t)
    r2s = r2_score(y_test, test_pred)
    print("R2 for test:", r2s)
    train_mse = mean_squared_error(train_pred, y_train)
    #print ('Train MSE: {}'.format(train_mse))
    test_mse = mean_squared_error(test_pred, y_test)
    #print ('Test MSE: {}'.format(test_mse))
    train_rmse = math.sqrt(train_mse)
    print("Train RMSE:", train_rmse)
    test_rmse = math.sqrt(test_mse)
    print("Test RMSE:", test_rmse)
    train_mae = mean_absolute_error(y_train, train_pred)
    #print("Train MAE:", train_mae)
    test_mae = mean_absolute_error(y_test, test_pred)
    #print("Test MAE:", test_mae)
```

```python
def RegModelScoresL(X_train_scaled, X_test_scaled, y_train, y_test):

    from sklearn import linear_model
    lm = linear_model.LinearRegression()
    lm.fit(X_train_scaled,y_train)

    import math
    from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
    train_pred = lm.predict(X_train_scaled)
    test_pred = lm.predict(X_test_scaled)
    r2t = r2_score(y_train, train_pred)
    print("R2 for train:", r2t)
    r2s = r2_score(y_test, test_pred)
    print("R2 for test:", r2s)
    train_mse = mean_squared_error(train_pred, y_train)
    #print ('Train MSE: {}'.format(train_mse))
    test_mse = mean_squared_error(test_pred, y_test)
    #print ('Test MSE: {}'.format(test_mse))
    train_rmse = math.sqrt(train_mse)
    print("Train RMSE:", train_rmse)
    test_rmse = math.sqrt(test_mse)
    print("Test RMSE:", test_rmse)
    train_mae = mean_absolute_error(y_train, train_pred)
    #print("Train MAE:", train_mae)
    test_mae = mean_absolute_error(y_test, test_pred)
    #print("Test MAE:", test_mae)
```

```python
def RegModelScoresKNN(X_train_scaled, X_test_scaled, y_train, y_test):

    from sklearn.neighbors import KNeighborsClassifier
    knn = KNeighborsClassifier()
    knn.fit(X_train_scaled, y_train)
    
    y_pred = knn.predict(X_test_scaled)
    from sklearn.metrics import confusion_matrix
    confusion_matrix(y_test, y_pred)
    from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
    train_pred = knn.predict(X_train_scaled)
    test_pred = knn.predict(X_test_scaled)
    r2t = r2_score(y_train, train_pred)
    print("R2 for train:", r2t)
    r2s = r2_score(y_test, test_pred)
    print("R2 for test:", r2s)
    train_mse = mean_squared_error(train_pred, y_train)
    #print ('Train MSE: {}'.format(train_mse))
    test_mse = mean_squared_error(test_pred, y_test)
    #print ('Test MSE: {}'.format(test_mse))
    train_rmse = math.sqrt(train_mse)
    print("Train RMSE:", train_rmse)
    test_rmse = math.sqrt(test_mse)
    print("Test RMSE:", test_rmse)
    train_mae = mean_absolute_error(y_train, train_pred)
    #print("Train MAE:", train_mae)
    test_mae = mean_absolute_error(y_test, test_pred)
    #print("Test MAE:", test_mae)
```
